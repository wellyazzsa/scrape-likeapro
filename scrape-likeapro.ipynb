```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.gunungbagging.com/prominence/'
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
soup.find('table')
```

```python
header = soup.find_all('th')
header_table = [title.text.strip() for title in header[:-1]]

header_table
```

```python
import pandas as pd

df = pd.DataFrame(columns=header_table)
df
```

```python
column = soup.find_all('tr')

for row in column[1:]:
    content = row.find_all('td')
    content_data = [data.text.strip() for data in content]

    del content_data[6]

    length = len(df)
    df.loc[length] = content_data

df
```

```python
df.to_csv(r',mountain.csv')
```
